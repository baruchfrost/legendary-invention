```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

spx = pd.read_csv('spx500_monthly_data.csv')

# Filter the data to only include years 2015-2018
spx = spx[(spx['date'] >= '2015-01-01') & (spx['date'] <= '2018-12-31')]

spx = spx.pivot(index='date', columns='ticker')
spx = spx.fillna(0)
spx.columns = spx.columns.droplevel()
spx.index = pd.to_datetime(spx.index)
spx = spx.reset_index()
spx = spx.groupby(pd.PeriodIndex(spx['date'],freq="M")).mean()
```

```python
# Plot all the values
plt.figure(figsize=(12, 6))
spx.plot()
plt.show()
```
